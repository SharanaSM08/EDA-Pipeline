```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display first rows
print(df.head())
```

```python
# Shape of dataset
print("Shape:", df.shape)

# Info
print(df.info())

# Check missing values
print("Missing Values:\n", df.isnull().sum())

# Statistical summary
print(df.describe())

# Check duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Remove duplicate rows
df_clean = df.drop_duplicates()

# Handle missing values (if any)
df_clean = df_clean.fillna(df_clean.mean(numeric_only=True))

# Convert categorical column 'type' into numeric
# The 'type' column does not exist in the California housing dataset.
# df_clean['type'] = df_clean['type'].map({'white': 0, 'red': 1})

# Reset index
df_clean.reset_index(drop=True, inplace=True)

print("Cleaning Done!")
```

```python
print("New Shape:", df_clean.shape)

print("Missing Values After Cleaning:\n", df_clean.isnull().sum())

print("Duplicates After Cleaning:", df_clean.duplicated().sum())
```

```python
df_clean.hist(figsize=(12,10))
plt.show()
```

```python
plt.figure(figsize=(10,6))
df_clean.boxplot()
plt.xticks(rotation=45)
plt.show()
```

```python
df_clean['housing_median_age'].value_counts().sort_index().plot(kind='bar')
plt.title("Housing Median Age Distribution")
plt.xlabel("Housing Median Age")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm')
plt.show()
```
